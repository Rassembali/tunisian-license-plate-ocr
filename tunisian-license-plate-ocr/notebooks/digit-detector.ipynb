```python
import os
import random
import shutil
from pathlib import Path

base_dir = '/kaggle/input/croped-licence-plat/train'
output_dir = '/kaggle/working/char_dataset_split'
train_dir = os.path.join(output_dir, 'train')
valid_dir = os.path.join(output_dir, 'valid')

# Create target dirs
for split in [train_dir, valid_dir]:
    os.makedirs(os.path.join(split, 'images'), exist_ok=True)
    os.makedirs(os.path.join(split, 'labels'), exist_ok=True)

# Get all images
images = list(Path(base_dir, 'images').glob('*.jpg'))
random.shuffle(images)

# 80/20 split
split_index = int(len(images) * 0.8)
train_images = images[:split_index]
valid_images = images[split_index:]

# Helper to move files
def move_files(image_list, split_path):
    for img_path in image_list:
        label_path = Path(base_dir, 'labels', img_path.stem + '.txt')
        shutil.copy(img_path, Path(split_path, 'images', img_path.name))
        shutil.copy(label_path, Path(split_path, 'labels', label_path.name))

move_files(train_images, train_dir)
move_files(valid_images, valid_dir)

print(f"âœ… Split complete. Train: {len(train_images)} | Valid: {len(valid_images)}")
```

```python
import os
import random
import shutil
from pathlib import Path

# Original data
base_images = '/kaggle/input/croped-licence-plat/train/images'
base_labels = '/kaggle/input/croped-licence-plat/train/labels'

# Target output
output_base = '/kaggle/working/char_dataset_split'
train_images_out = os.path.join(output_base, 'train/images')
train_labels_out = os.path.join(output_base, 'train/labels')
valid_images_out = os.path.join(output_base, 'valid/images')
valid_labels_out = os.path.join(output_base, 'valid/labels')

# Create folders
for folder in [train_images_out, train_labels_out, valid_images_out, valid_labels_out]:
    os.makedirs(folder, exist_ok=True)

# Gather and shuffle
all_images = list(Path(base_images).glob('*.jpg'))
random.shuffle(all_images)

# Split
split_index = int(0.8 * len(all_images))
train_images = all_images[:split_index]
valid_images = all_images[split_index:]

# Copy function
def copy_files(images, img_out, lbl_out):
    for img_path in images:
        label_path = Path(base_labels) / (img_path.stem + '.txt')
        shutil.copy(img_path, img_out)
        shutil.copy(label_path, lbl_out)

copy_files(train_images, train_images_out, train_labels_out)
copy_files(valid_images, valid_images_out, valid_labels_out)

print(f"âœ… Split done: Train={len(train_images)}, Valid={len(valid_images)}")
```

```python
from ultralytics import YOLO

# Load the base YOLOv8 model
model = YOLO('yolov8n.pt')

# Train using your dataset
model.train(
    data='/kaggle/working/char_dataset_split/data.yaml',
    epochs=30,
    imgsz=640,
    batch=16
)
```
